```python
fv_nodes  = repograph["function_version_nodes"]
fv_edges  = repograph["function_version_edges"]
fv_func   = repograph["functionversion_function_edges"]
commit_df = repograph["commit_nodes"]
fv_cmt    = repograph["functionversion_commit_edges"]

print("FV nodes:", len(fv_nodes),
      "| NEXT_VERSION edges:", len(fv_edges),
      "| Version→Function edges:", len(fv_func),
      "| Commits:", len(commit_df),
      "| Version→Commit edges:", len(fv_cmt))

display(fv_nodes.head(10))
display(commit_df.head(5))
```

```python
# 5/a) keressünk egy olyan symbol_id-t, amin több verzió is van
multi_chain = (
    fv_nodes.groupby("symbol_id")["ID"].count()
    .reset_index()
    .query("ID >= 2")
    .sort_values("ID", ascending=False)
)

if not multi_chain.empty:
    symbol = multi_chain.iloc[0]["symbol_id"]
    print("Példa symbol_id:", symbol)

    chain = (fv_nodes[fv_nodes["symbol_id"] == symbol]
             .sort_values(["authored_datetime", "commit_sha"]))
    display(chain[["ID","symbol_id","commit_sha","authored_datetime","start_line","end_line"]])
else:
    print("Nem találtam legalább 2 verziós függvényt a kiválasztott részfában.")
```

```python
import networkx as nx
from pyvis.network import Network

def visualize_version_chain(fv_nodes_df, fv_edges_df, symbol_id, html_out="version_chain.html"):
    sub_nodes = fv_nodes_df[fv_nodes_df["symbol_id"] == symbol_id].copy()
    sub_nodes = sub_nodes.sort_values(["authored_datetime","commit_sha"])
    keep_ids = set(sub_nodes["ID"].tolist())
    sub_edges = fv_edges_df[fv_edges_df["source"].isin(keep_ids) & fv_edges_df["target"].isin(keep_ids)].copy()

    G = nx.DiGraph()
    for _, r in sub_nodes.iterrows():
        label = f"v{r['ID']}\n{r['commit_sha'][:7]}\n{(r['authored_datetime'] or '')[:19]}"
        G.add_node(int(r["ID"]), label=label, title=r.get("qualified_name",""), color="#1a70aa")

    for _, r in sub_edges.iterrows():
        G.add_edge(int(r["source"]), int(r["target"]), color="#a6cee3")

    net = Network(height='700px', width='100%', directed=True, notebook=False)
    net.from_nx(G)
    net.show_buttons(filter_=['physics'])
    net.save_graph(html_out)
    print(f"Lánc vizualizáció mentve ide: {html_out}")

# futtatás az előbb talált symbol_id-ra (ha volt)
if 'symbol' in locals():
    visualize_version_chain(fv_nodes, fv_edges, symbol, html_out="version_chain.html")
```

```python
# A KGB az alábbi helyre klónoz: ./repos/<repo_name>
# A repo ROOT a részfa előtt van; ha részfát adtál meg, vissza tudod számolni:
repo_root = os.path.join("./repos", "scikit-learn")  # a klónozás neve

# A CG node-ok a repograph-ban:
cg_nodes = repograph["function_nodes"].rename(columns={"ID":"func_id"})  # a FV úgyis kezeli, de egyértelműbb

# Saját FV példány korlátozott commit-számmal
fv = FunctionVersioning(repo_path=repo_root, log_errors=True)
fv_out = fv.build(
    function_nodes_df=cg_nodes,
    branch_ref="HEAD",
    only_changed_files=True,
    max_commits=250,     # pl. az utolsó 250 commit
)

# Eredmények
fv_nodes2  = fv_out["function_version_nodes"]
fv_edges2  = fv_out["function_version_edges"]
fv_func2   = fv_out["functionversion_function_edges"]
commit2    = fv_out["commit_nodes"]
fv_cmt2    = fv_out["functionversion_commit_edges"]

print("FV nodes:", len(fv_nodes2),
      "| NEXT_VERSION edges:", len(fv_edges2),
      "| Version→Function edges:", len(fv_func2),
      "| Commits:", len(commit2),
      "| Version→Commit edges:", len(fv_cmt2))

display(fv_nodes2.head(10))
```
